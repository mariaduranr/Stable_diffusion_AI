## Git

Git is a command line tool for working with git repositories.

Github is probably the best known site for storing git repositories.  There is an application called Github Desktop that provides a graphical user interface for interacting with repositories on the github site. Since we need to download repositories to lambda1, a linux machine, we will use the command line tool git.

git is a very powerful tool for sharing code and maintaining different versions of a project.

Because we will only be downloading and updating repositories, we need only three git tools:

>git clone
>
>git pull
>
>git status

### Webui

When I setup your accounts on lambda1, I included a version of the git repository for stable-diffusion-webui. I did that so I could edit the individual webui-user.sh and give all of you a port number.

I used the following command to clone the stable-diffusion-webui repository in my home directory:

>git clone https://github.com/AUTOMATIC1111/stable-diffusion-webui.git

This create the directory stable-diffusion-webui in my home directory.

### Dreambooth Extension

You can use the GUI webui to install the dreambooth extension, but you can also install it using got clone. Here are the steps:

>cd stable-diffusion-webui/extensions
>
>git clone https://github.com/d8ahazard/sd_dreambooth_extension.git

This will create the folder sd_dreambooth_extension in stable-diffusion-webui/extensions folder.

After creating the directory you need to restart webui.

### git status

You can check to see if you have the latest version of the repository with the git status command:
