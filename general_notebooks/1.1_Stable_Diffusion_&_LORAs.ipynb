# Stable Diffusion

## Introduction to Stable Diffusion

**What is it?**
    Stable Diffusion is an artificial intelligence (AI) tool for image generation that was released open source on August 22, 2022 by StabilityAI. Stable Diffusion has risen in popularity due to it being open source, because it is free, and because it can be run on an ordinary computer, and several other image generation tools have been built on top of Stable Diffusion models. 
    
**Uses of Stable Diffusion**
    
* Text-to-Image Generation: Enter prompts, ideally both positive (aspects you do want) and negative (aspects you don't want), and receive a generated image. 
    * This is possible in many different styles, including realistic, semi-realistic, anime, game assets, landscapes, etc.)
   
    
* Image-to-Image Generation: Use existing images as inspiration to create new images or transform them. 
    * Fix and edit images using inpainting (painting over a specific part and regenerate the image) and/or outpainting (extending an image by painting/generating outside the frame).
    
    
* Create animations
    
Read more at: https://aituts.com/stable-diffusion-start and https://sandner.art/ultimate-guide-to-sd-15-photorealism/

# LORAs

Here's some terminology that will be helpful throughout our pre-prac

CHECKPOINTS (GB)

A "checkpoint" refers to a saved state of the full model that captures all the information necessary to reproduce the model's operations from a particular point in time. Serves as starting points for the generation of images.

Storing:
<br>-weights (the parameters that have been learned during training)
<br>-architecture (the structure of the model, including the layers and their connections)
<br>-other metadata related to the model's training process

These are more for handling broad concepts, like images that are more realistic or more cartoon like, rather than creating something more specific, like generating consistent pictures of a beagle.

Add: Checkpoint vs SafeTensors

LORA (MB)

"Low-Rank Adaptation" models are pre-trained with minimal additional parameters making it lightweight. Reducing dimensionality to produce specific features and themes that achieves better performance on specific tasks or exhibit certain stylistic tendencies. Customization of the model's behavior without the need for extensive retraining on data

These are more for generating faces, objects, or general environments

EMBEDDINGS

A representation of data (images, words, or other) in a continuous vector space. Vector representations capture the essential qualities or features in a form that the model can process. Representing different styles, concepts, or features that can be incorporated into the image generation process. These features can be in place of a image prompt.

These are more for editing and be used in either positive or negative prompts
